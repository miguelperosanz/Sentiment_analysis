# SCRIPT FOR SCRAPING THE "INCOMPLETE" TWITTER ACCOUNTS

```python
import pandas as pd


# Numbers from the "incompletely downloaded" twitter accounts:


k=[16,26,41,45,70,89,92,94,100,116,141,142,152,154,168,170,171,172,179,183,189,199,215,230,236,240,257,264,273,278,287,294]


for i in range(len(k)):
    

    
    a = str(k[i])
    
    p = open("./scrap_twitter_accounts/script.sh","a+")
    
    file=str("./scrap_twitter_accounts/output_got"+a+".csv")
    
    f = open(file, "r")

    data = pd.read_csv(f) 
    data['date'] = pd.to_datetime(data['date'],format="%Y-%m-%d")
    data['date'] = data['date'].dt.date


    list_names=data['username']
    min_date=data['date'].min()


 
    
    print('GetOldTweets3 --username "',list_names[0],'" --since 2018-01-01 ',' --until ',min_date - pd.Timedelta(days=1),sep='',file=p)
    print('mv output_got.csv output_got',k[i],'_b.csv',sep='',file=p)
    print('sleep 300',file=p)
```
