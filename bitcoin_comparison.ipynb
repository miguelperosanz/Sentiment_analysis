# LET'S STUDY NOW "EXTREM" TWEETS: TWEETS OBVIOUSLY GOOD OR OBVIOUSLY BAD.


```python
import pandas as pd


tweets_good = pd.read_csv('./good_tweets.csv')
tweets_bad = pd.read_csv('./bad_tweets.csv')
```

```python
tweets_good
```

```python
tweets_bad
```

LET'S START WITH THE POSITIVE TWEETS

```python
one_star_good = tweets_good['1_stars'].sum()
two_stars_good = tweets_good['2_stars'].sum()
three_stars_good= tweets_good['3_stars'].sum()
four_stars_good = tweets_good['4_stars'].sum()
five_stars_good = tweets_good['5_stars'].sum()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
fig= plt.figure(figsize=(14,7))

height = [one_star_good, two_stars_good, three_stars_good, four_stars_good, five_stars_good]
bars = ('1 star', '2 stars', '3 stars', '4 stars', '5 stars')
y_pos = np.arange(len(bars))
 

plt.bar(y_pos, height, color = (0.5,0.1,0.5,0.6))

plt.title('Model applied to the set of positive tweets, adding the rates for every star with .sum')
plt.xlabel('categories')
plt.ylabel('added total value for every star')
 

plt.ylim(0,14)
 

plt.xticks(y_pos, bars)

plt.show()
```

```python
one_star_good2 = len(tweets_good[tweets_good['winning_star']==1])
two_stars_good2 = len(tweets_good[tweets_good['winning_star']==2])
three_stars_good2 = len(tweets_good[tweets_good['winning_star']==3])
four_stars_good2 = len(tweets_good[tweets_good['winning_star']==4])
five_stars_good2 = len(tweets_good[tweets_good['winning_star']==5])
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
fig= plt.figure(figsize=(14,7))

height = [one_star_good2, two_stars_good2, three_stars_good2, four_stars_good2, five_stars_good2]
bars = ('1 star', '2 stars', '3 stars', '4 stars', '5 stars')
y_pos = np.arange(len(bars))
 

plt.bar(y_pos, height, color = (0.5,0.1,0.5,0.6))

plt.title('Model applied to the set of positive tweets, just counting how many times every star "won"')
plt.xlabel('categories')
plt.ylabel('Number of times that every label "won"')
 

plt.ylim(0,18)
 

plt.xticks(y_pos, bars)

plt.show()
```
